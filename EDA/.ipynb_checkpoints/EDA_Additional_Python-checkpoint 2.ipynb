```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fifa_data = pd.read_excel(r'fifa_data_for_eda.xlsx')
```

```python
fifa_data.head()
```

```python
pd.DataFrame(fifa_data.columns)
```

```python
fifa_data.dtypes
```

```python
fifa_data_copy=fifa_data.copy()
```

```python
x=['S.No.','sofifa_id']
y=['nationality','Player_Continent','Club_Continent','Club_Int_Prestige','Club_Domestic_Prestige','preferred_foot'
  ,'Player_international_reputation','weak_foot','Position_Final']
```

```python
fifa_data_copy[x] = fifa_data_copy[x].astype('str')
fifa_data_copy[y] = fifa_data_copy[y].astype('category')
```

```python
fifa_data_copy.info()
```

```python
categorical_data=fifa_data_copy.select_dtypes('category')
```

```python
categorical_data=categorical_data.iloc[:,1:]
```

```python
categorical_data.columns[4]
```

```python
sns.set(sns.set_context('poster',font_scale=1.7))
fig,axes=plt.subplots(ncols=1,nrows=4,figsize=(25,40))
plt.ticklabel_format(useOffset=False, style='plain')
for i in range (0,4):
    sns.boxplot(y=categorical_data.columns[i],x='Value',data=fifa_data_copy,ax=axes[i],orient='h')
```

```python
from matplotlib import rcParams
rcParams['axes.titlepad'] = 20 
sns.set(sns.set_context('poster',font_scale=1.7))
fig,axes=plt.subplots(ncols=1,nrows=4,figsize=(25,40))
plt.ticklabel_format(useOffset=False, style='plain')
i=4
while i<7:
    g=sns.boxplot(y=categorical_data.columns[i],x='Value',data=fifa_data_copy,ax=axes[i-4])
    plt.ylabel(categorical_data.columns[i], labelpad=10)
    i=i+1
    
while i==7:
    g=sns.boxplot(y=categorical_data.columns[i],x='Value',data=fifa_data_copy,ax=axes[i-4])
    i=i+1
    plt.setp(g.get_yticklabels(), rotation=30)
```

```python
sns.set(sns.set_context('poster',font_scale=1.3))
fig,axes=plt.subplots(ncols=1,nrows=4,figsize=(25,40))
plt.ticklabel_format(useOffset=False, style='plain')
for i in range (0,4):
    sns.boxplot(y=categorical_data.columns[i],x='Wages',data=fifa_data_copy,ax=axes[i])
```

```python
sns.set(sns.set_context('poster',font_scale=2))
fig,axes=plt.subplots(ncols=1,nrows=4,figsize=(30,60))
plt.ticklabel_format(useOffset=False, style='plain')
for i in range (4,8):
    g=sns.boxplot(y=categorical_data.columns[i],x='Wages',data=fifa_data_copy,ax=axes[i-4])
    plt.setp(g.get_yticklabels(), rotation=30)
```

```python
numerical_data = fifa_data_copy.select_dtypes(np.number)
```

```python
numerical_data.drop(['Value','Wages'],axis=1,inplace=True)
```

```python
len(numerical_data.columns)
```

```python
sns.set(sns.set_context('paper',font_scale=2))
fig,axes=plt.subplots(ncols=3,nrows=5,figsize=(20,30))
plt.ticklabel_format(useOffset=False, style='plain')
k=0
plt.tight_layout()
for i in range (0,5):
    for j in range (0,3):
        k1=sns.boxplot(x=numerical_data.columns[k],y='Player_Continent',data=fifa_data_copy,ax=axes[i,j])
        if k%3 !=0 :
            k1.axes.get_yaxis().set_visible(False)
        else:
            pass
        k+=1
```
